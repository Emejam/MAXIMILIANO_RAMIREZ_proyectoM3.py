```python
import numpy as np  #AQUI IMPORTAMOS LA BIBLIOTECA PARA UTILIZAR LA FUNCION RANDOM CHOICE
np.random.choice ([-1, 1])
```

```python
import matplotlib.pyplot as plt #AQUI IMPORTAMOS LA BIBLIOTECA PARA GRAFICAR NUESTROS RESULTADOS
```

```python
def maquina_de_galton(n_pelotas, n_niveles): #DEFINIMOS NUESTRA FUNCION MAQUINA DE GALTON
  resultados = []  #SE CREA UNA LISTA PARA GUARDAR NUESTROS RESULTADOS

  for i in range(n_pelotas):
    posicion = 0  #SE COMIENZA EN LA POSICION 0
    for i in range(n_niveles):
      movimiento = np.random.choice([-1,1]) #LE DAMOS COMO OPCION EL SUMAR UN VALOR O RESTAR UN VALOR
      posicion += movimiento
    resultados.append(posicion)  #SE GUARDA NUESTRO RESULTADO EN LA LISTA
  print(resultados) #SE IMPRIMEN LOS RESULTADOS
  return resultados
```

```python
def tabla(resultados, n_niveles): # DEFINIMOS LA FUNCION TABLA QUE NOS VA A SERVIR PARA GRAFICAR LOS RESULTADOS
  plt.hist(resultados, bins=2*n_niveles + 1, edgecolor = 'grey') #SE DEFINEN LOS PARAMETROS, NOMBRES Y COLORES DE LA GRAFICA
  plt.title("Maquina de Galton")
  plt.xlabel("Posicion")
  plt.ylabel("Frecuencia")
  plt.grid(True)
  plt.show()
```

```python
n_pelotas = 3000 #DEFINIMOS LOS PARAMETROS DE NUMERO DE PELOTAS Y MNUMERO DE NIVELES
n_niveles =12


resultados = maquina_de_galton (n_pelotas, n_niveles)
tabla(resultados, n_niveles)
```
